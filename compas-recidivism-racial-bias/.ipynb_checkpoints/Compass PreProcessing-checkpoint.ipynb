```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
RawScore_df = pd.read_csv('compas-scores-raw.csv')
```

```python
RawScore_df.info()
```

```python
RawScore_df.head(100)
```

### Change Date of Birth to 4 digit year ### 

```python
RawScore_df['DateOfBirth']=RawScore_df['DateOfBirth'].apply(lambda x: str(x[:5]) +'/'+'19'+str(x[6:]))
```

```python
RawScore_df['DateOfBirth']=pd.to_datetime(RawScore_df['DateOfBirth'], format='%m/%d/%Y') #convert to datetime object
```

```python
RawScore_df['BirthYear']=RawScore_df['DateOfBirth'].dt.year #get the year for BirthDate
```

```python
RawScore_df['Screening_Date']=pd.to_datetime(RawScore_df['Screening_Date'], infer_datetime_format=True) #convert to datetime object
```

```python
RawScore_df['ScreeningYear']=RawScore_df['Screening_Date'].dt.year #get the year for ScreeningDate
```

## Age Variable ##

```python
RawScore_df['Age']=(RawScore_df['ScreeningYear']-RawScore_df['BirthYear'])
```

```python
RawScore_df = RawScore_df.loc[RawScore_df['BirthYear']>=1945] #removes outliers
```

```python
RawScore_df['Age'].describe()
```

## Make Seasons Variable ##
    

```python
RawScore_df['ScreeningMonth']=RawScore_df['Screening_Date'].dt.month #get the month for ScreeningDate
```

```python
RawScore_df['ScreenSeason'] = 'All' #creates ScreenSeason column
RawScore_df.loc[RawScore_df['ScreeningMonth'] == 12, 'ScreenSeason'] = 'Winter' 
RawScore_df.loc[(RawScore_df['ScreeningMonth'] >=1) & (RawScore_df['ScreeningMonth'] <3), 'ScreenSeason'] = 'Winter'
RawScore_df.loc[(RawScore_df['ScreeningMonth'] >=3) & (RawScore_df['ScreeningMonth'] <6), 'ScreenSeason'] = 'Spring'
RawScore_df.loc[(RawScore_df['ScreeningMonth'] >=6) & (RawScore_df['ScreeningMonth'] <9), 'ScreenSeason'] = 'Summer'
RawScore_df.loc[(RawScore_df['ScreeningMonth'] >=9) & (RawScore_df['ScreeningMonth'] <12), 'ScreenSeason'] = 'Fall'
```

## Sex_Code_Text Feature (binary): 0=Female, 1=Male

```python
RawScore_df['Sex_Code_Text'].replace('Female', 0,inplace=True)
RawScore_df['Sex_Code_Text'].replace('Male', 1,inplace=True)
```

```python
RawScore_df.head()
```

## Ethnic_Code_Text (one-hot)

```python
RawScore_df = pd.concat([RawScore_df.drop('Ethnic_Code_Text', axis=1), pd.get_dummies(RawScore_df['Ethnic_Code_Text'])], axis=1)
```

```python
# renaming Other column name to avoid naming conflicts
RawScore_df=RawScore_df.rename(columns = {'Other':'Other_ethnicity'})
```

```python
RawScore_df.head()
```

## Language(binary): 0=English 1=Spanish

```python
RawScore_df['Language'].replace('English', 0, inplace=True)
RawScore_df['Language'].replace('Spanish', 1, inplace=True)
```

## LegalStatus (one-hot)

```python
RawScore_df = pd.concat([RawScore_df.drop('LegalStatus', axis=1), pd.get_dummies(RawScore_df['LegalStatus'])], axis=1)
```

```python
# renaming Other column name to avoid naming conflicts
RawScore_df=RawScore_df.rename(columns = {'Other':'Other_legal'})
```

```python
RawScore_df.head()
```

## CustodyStatus (one-hot)

```python
RawScore_df = pd.concat([RawScore_df.drop('CustodyStatus', axis=1), pd.get_dummies(RawScore_df['CustodyStatus'])], axis=1)
```

```python
RawScore_df.head()
```

## MaritalStatus (one-hot)

```python
RawScore_df = pd.concat([RawScore_df.drop('MaritalStatus', axis=1), pd.get_dummies(RawScore_df['MaritalStatus'])], axis=1)
```

```python
RawScore_df.head()
```
