```python
import pandas as pd
import numpy as np
```

```python
RawScore_df = pd.read_csv('compass/compas-scores-raw.csv')
```

```python
list_of_features=RawScore_df.DisplayText.unique()
list_of_features
```

```python
for i in list_of_features: # for each of the elements given in the dataframe
    RawScore_df[i]=[RawScore_df.iloc[idx,22] if v==i else np.nan for idx,v in enumerate(RawScore_df['DisplayText'])] #creates a feature for each element. 
```

```python
RawScore_df.head(100)
```

```python
from datetime import datetime
RawScore_df['DateOfBirth']=datetime.strptime(RawScore_df['DateOfBirth'], '%m/%d/%y')
```

```python
RawScore_df['DateOfBirth']=pd.to_datetime(RawScore_df['DateOfBirth'], format='%m/%d/%y')
```

```python
RawScore_df['DateOfBirth'].describe()
```

```python
RawScore_df['BirthYear']=RawScore_df['DateOfBirth'].dt.year
```

```python
np.where(RawScore_df['BirthYear']>2015)
```

```python
RawScore_df['Screening_Date'].describe()
```

```python
RawScore_df['Screening_Date']=pd.to_datetime(RawScore_df['Screening_Date'].astype(str), format='%m/%d/%y')
```

```python
RawScore_df['ScreeningYear']=RawScore_df['Screening_Date'].dt.year
```

```python
RawScore_df['Age']=(RawScore_df['ScreeningYear']-RawScore_df['BirthYear'])
```
