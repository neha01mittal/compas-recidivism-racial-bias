### Change Date of Birth to 4 digit year ### 

```python
RawScore_df['DateOfBirth']=RawScore_df['DateOfBirth'].apply(lambda x: str(x[:5]) +'/'+'19'+str(x[6:]))
```

```python
RawScore_df['DateOfBirth']=pd.to_datetime(RawScore_df['DateOfBirth'], format='%m/%d/%Y') #convert to datetime object
```

```python
RawScore_df['BirthYear']=RawScore_df['DateOfBirth'].dt.year #get the year for BirthDate
```

```python
RawScore_df['Screening_Date']=pd.to_datetime(RawScore_df['Screening_Date'], infer_datetime_format=True) #convert to datetime object
```

```python
RawScore_df['ScreeningYear']=RawScore_df['Screening_Date'].dt.year #get the year for ScreeningDate
```

## Age Variable ##

```python
RawScore_df['Age']=(RawScore_df['ScreeningYear']-RawScore_df['BirthYear'])
```

```python
RawScore_df = RawScore_df.loc[RawScore_df['BirthYear']>=1945] #removes outliers
```
